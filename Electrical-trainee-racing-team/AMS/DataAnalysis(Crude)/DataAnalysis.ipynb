```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
df = pd.read_csv("AMSCelldata97 (1).csv")

# Time array (100 ms samples)
time = df.index * 0.1

# TSV vs Time
plt.figure(figsize=(10, 4))
plt.plot(time, df['TSV'], label='TSV (V)', color='blue')
plt.scatter(time, df['TSV'], marker='.', s=5)
plt.xlabel('Time (s)')
plt.ylabel('TSV (V)')
plt.title('Total Stack Voltage vs Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# TSC vs Time
plt.figure(figsize=(10, 4))
plt.plot(time, df['TSC'], label='TSC (A)', color='green')
plt.scatter(time, df['TSC'], marker='.', s=5)
plt.xlabel('Time (s)')
plt.ylabel('TSC (A)')
plt.title('Total Stack Current vs Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#Average
average_values = df.iloc[:, 0:95].mean()

plt.figure(figsize=(8, 5))
average_values.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Cell Voltages')
plt.ylabel('Voltage Readings (V)')
plt.xlabel('')          
plt.xticks([]) 
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Power
power = df['TSV'] * df['TSC']
average_power = np.mean(power[power != 0])
print(f'Average Power: {np.abs(average_power):.2f}')
```
